```python
df.loc[len(df.index)] = ['is it possible to get rid of diabetes', 'There is no known cure for type 2 diabetes. But it can be controlled. And in some cases, it goes into remission. For some people, a diabetes-healthy lifestyle is enough to control their blood sugar levels.'] 
```

```python
df
```

```python
Questions_list = []
Answers_list = []
for q in df['Questions']:Questions_list.append(q)
for q in df['Answers']:Answers_list.append(q)
for i in Questions_list:i.replace('\xa0',' ') 
for i in Answers_list:i.replace('\xa0',' ') 
Questions_list = iter(Questions_list)
Answers_list = iter(Answers_list)
```

```python
n = 0
for q  in range(len(df['Questions'])):
    tag = 'diabetes' + str(n)
    n = n + 1
    data['intents'].append({'tag': tag,
    'patterns': [next(Questions_list)],
    'responses': [next(Answers_list)],
    'context_set': ''})
    
```
